#### rajout d'une colonne catégorie

```python
def assign_categorie(type_equipement):
    if type_equipement in ['Bibliothèque', "Centre d'art", 'Centre de création artistique','théatre', 'cinéma','Centre de création musicale', 'Théâtre', 'Conservatoire','Parc et jardin','Librairie', 'Lieu de mémoire', 'Musée', 'Opéra',"Centre culturel", "Scène", "Espace protégé"]:
        return ('vivant')
    elif type_equipement in ["Service d'archives", 'Monument']:
        return ('patrimoine')
    elif type_equipement ==  "Établissement d'enseignement supérieur":
        return ('Strange')
    else:
        return None  # Si un monument ne correspond à aucune catégorie
```

```python
# Appliquer la fonction à la colonne 'monument'
df_culturels['categorie'] = df_culturels['Type équipement ou lieu'].apply(assign_categorie)
df_culturels.head()
```

### paramètres utiles

```python
param_utiles=['Nom','Adresse','Code Postal','code_insee','libelle_geographique','Type équipement ou lieu','categorie','Latitude','Longitude']
param_utiles
```

```python
df_culturels_extracted = df_culturels.loc[:,param_utiles]
df_culturels_extracted.head()
```

```python
df_culturels_extracted.groupby('code_insee')
```

```python
# Regroupement par 'code_postal' et 'categorie', puis comptage des éléments
df_nbre_categorie_parcommune = df_culturels_extracted.groupby(['code_insee', 'categorie']).size().reset_index(name='count')
df_nbre_categorie_parcommune
```

```python
df_nbre_categorie_parcommune.to_csv(path_data+'nbre_categorie_parcommune.csv', sep='\t', index=False)  # Save as a tab
```

```python
# Regroupement par 'code_postal' et 'type equipement', puis comptage des éléments
df_nbre_typequip_parcommune = df_culturels_extracted.groupby(['code_insee', 'Type équipement ou lieu']).size().reset_index(name='count')
df_nbre_typequip_parcommune
```

```python
df_nbre_typequip_parcommune.to_csv(path_data+'nbre_typequip_parcommune.csv', sep='\t', index=False)  # Save as a tab
```

```python
# Ajouter la colonne Code postal, en utilisant 'transform' pour ne pas perdre l'information
df_culturels_extracted['Code postal'] = df_culturels_extracted.groupby('code_insee')['Code Postal'].transform('first')

# Regrouper par Code commune et Catégorie, puis compter les éléments
grouped = df_culturels_extracted.groupby(['Code postal', 'code_insee', 'categorie']).size().reset_index(name='Nombre d\'éléments')

# Afficher le résultat final
print(grouped)
```
