## Libraries

```python
import pandas as pd
import numpy as np
```

```python
def assign_categorie(type_equipement):
    if type_equipement in ['Bibliothèque', "Centre d'art", 'Centre de création artistique', 'Cinéma','Centre de création musicale', 'Théâtre', 'Conservatoire','Librairie', 'Musée', 'Opéra',"Centre culturel", "Scène"]:
        return ('vivant')
    elif type_equipement in ["Service d'archives", 'Monument','Parc et jardin', 'Lieu de mémoire', "Espace protégé",'Lieu archéologique']:
        return ('patrimoine')
    elif type_equipement ==  "Établissement d'enseignement supérieur":
        return ('Strange')
    else:
        return None  # Si un monument ne correspond à aucune catégorie
```

### paramètres utiles

```python
param_utiles=['Nom','Adresse','Code Postal','code_insee','libelle_geographique','Nombre_de_salles_de_theatre','Type équipement ou lieu','Latitude','Longitude']
param_utiles
```

```python
df_culturels_extracted = df_culturels.loc[:,param_utiles]
df_culturels_extracted.head()
```

```python
# I- Regroupement par 'code_postal' et 'type equipement', puis comptage des éléments
df_nbre_typequip_parcommune = df_culturels_extracted.groupby(['code_insee', 'Type équipement ou lieu']).size().reset_index(name='count')
df_nbre_typequip_parcommune
```

```python
# II-Extraction de df avec le nbre de théatre par commune
df_nbre_theatre_par_commune = (df_culturels.loc[df_culturels['Type équipement ou lieu']=='Théâtre',['code_insee','Type équipement ou lieu','Nombre_de_salles_de_theatre']])
# calcul du nombre de theatre par commune
grouped2 = df_nbre_theatre_par_commune.groupby(['code_insee'])['Nombre_de_salles_de_theatre'].sum()
grouped2
```

```python
df_nbre_theatre_par_commune
```

```python
# III- creation d'une series avec le nombre de salle de theatre (index=code commune, valeurs=nbre de salles de theatres)
for i,id in enumerate(grouped2.index):    
    
    filtered_df = df_nbre_theatre_par_commune[(df_nbre_theatre_par_commune['code_insee'] == id) & (df_nbre_theatre_par_commune['Type équipement ou lieu'] == 'Théâtre')]

    s1 = filtered_df.groupby(['code_insee'])['Nombre_de_salles_de_theatre'].sum()
    if i==0:
        result_horizontal=s1
    else:
        result_horizontal = pd.concat([result_horizontal, s1], axis=0)

# V- mettre a jour le nbre de salles de theatre dans df_nbre_typequip_parcommune
for id in result_horizontal.index:
    #filtered_df = df_nbre_typequip_parcommune[(df_nbre_typequip_parcommune['code_insee'] == id) & (df_nbre_typequip_parcommune['Type équipement ou lieu'] == 'Théâtre')]
    mask=(df_nbre_typequip_parcommune['code_insee'] == id) & (df_nbre_typequip_parcommune['Type équipement ou lieu'] == 'Théâtre')
 
    df_nbre_typequip_parcommune.loc[mask,'count']=result_horizontal[id]
```

```python
# Appliquer la fonction à la colonne 'monument'
df_nbre_typequip_parcommune['categorie'] = df_nbre_typequip_parcommune['Type équipement ou lieu'].apply(assign_categorie)
df_nbre_typequip_parcommune
```
