```python
import pandas as pd
import sqlite3
```

```python
src_data_basilics = '../src/data_basilics/'
src_data_cinema = '../src/data_cinema/'
src_data_festivals = '../src/data_festivals/'
```

```python
festivals_data = pd.read_csv(src_data_festivals + 'festivals-global-festivals-_-pl.csv', delimiter=";")
basilics_data = pd.read_csv(src_data_basilics + 'base-des-lieux-et-des-equipements-culturels.csv', delimiter=";", low_memory=False)
```

```python
db_name = 'database.db' 
conn = sqlite3.connect(db_name)
```

```python
festivals_data_name = 'festivals_data'  
basilics_data_name = 'basilics_data'

festivals_data.to_sql(festivals_data_name, conn, if_exists='replace', index=False)
basilics_data.to_sql(basilics_data_name, conn, if_exists='replace', index=False)

#TODO cinema data



conn.commit()
conn.close()
```
