```python
import pandas as pd
import sqlite3
import os
```

```python
src_data_basilics = '../src/data_basilics/'
src_data_cinema = '../src/data_cinema/'
src_data_festivals = '../src/data_festivals/'
```

```python
festivals_data = pd.read_csv(src_data_festivals + 'festivals-global-festivals-_-pl.csv', delimiter=";")
basilics_data = pd.read_csv(src_data_basilics + 'base-des-lieux-et-des-equipements-culturels.csv', delimiter=";", low_memory=False)
villes_data= pd.read_csv('../src/' + 'programme-petites-villes-de-demain-liste-des-villes-beneficiaires.csv', delimiter=";", low_memory=False)
```

```python
db_name = 'database.db' 
conn = sqlite3.connect(db_name)
```

```python
festivals_data_name = 'festivals_data'  
basilics_data_name = 'basilics_data'
villes_data_name = 'villes_data'

festivals_data.to_sql(festivals_data_name, conn, if_exists='replace', index=False)
basilics_data.to_sql(basilics_data_name, conn, if_exists='replace', index=False)
villes_data.to_sql(villes_data_name, conn, if_exists='replace', index=False)

#TODO cinema data
#for year in range(2023, 2002, -1):
#    file_path = f"../src/data_cinema/extracted/Etablissements_{year}.xlsx"
#    df = pd.read_excel(file_path)
#    df['year'] = year
#    df.to_sql('etablissements_annual_data', conn, if_exists='append', index=False)


conn.commit()
conn.close()
```

```python
file_directory = "../src/data_cinema/extracted" 
excel_files = [f for f in os.listdir(file_directory) if f.endswith('.xlsx')]

unique_columns = set()
for file in excel_files:
    file_path = os.path.join(file_directory, file)
    df = pd.read_excel(file_path)
    unique_columns.update(df.columns)

unique_columns = sorted(list(unique_columns))
print(unique_columns)
```

```python
# Create the SQL command for creating the table
columns_sql = ",\n".join([f'"{col}" VARCHAR(255)' for col in unique_columns])

create_table_sql = f"""
CREATE TABLE etablissements_annual_data (
    year INT NOT NULL,
    {columns_sql}
);
"""

# Print the SQL command
print(create_table_sql)
```
